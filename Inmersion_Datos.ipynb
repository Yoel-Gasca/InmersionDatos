2- Reflegar el precio por metro cuadrado y reprensentarlo con un grafico que se adapte a esta demostracion.

```python
inmuebles['Precio_m2'] = inmuebles.Precio_Millon / inmuebles.Area
```

```python
datoStd = inmuebles.groupby(by = 'Barrio').describe()

topPromM2 = datoStd.Precio_m2[['mean', 'std']].sort_values(by= 'mean', ascending=False).head(10)
topPromM2 = topPromM2.reset_index()
```

```python
sns.set_theme()
plt.figure(figsize = (12,6))

sns.barplot(data = topPromM2, y = 'mean', x='Barrio')

sigma = topPromM2['std']

x = topPromM2['Barrio']
y = topPromM2['mean']

plt.errorbar(x = x, y = y, yerr=(sigma, sigma), fmt='none', color='k', capsize=5)

plt.title('Costo por m2 de inmuebles por barrio')
plt.xlabel('Barrio')
plt.ylabel('Promedio')

plt.xticks(rotation = 80)
plt.savefig('/content/drive/MyDrive/Inmersion-de-Datos-Alura/costoPorM2', format = 'png')
plt.show()
```

Primero de creo la columna con el valor por m2

Para despues agrupar y usar describe para obtener el promedio y la desviación estándar

Luego se generaria el grafico. Se uso barplot y tambien se agregaron barras de error usando la desviación estándar.

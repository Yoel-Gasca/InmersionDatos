```python
inmuebles['Precio_m2'] = inmuebles.Precio_Millon / inmuebles.Area
```

```python
datoStd = inmuebles.groupby(by = 'Barrio').describe()

topPromM2 = datoStd.Precio_m2[['mean', 'std']].sort_values(by= 'mean', ascending=False).head(10)
topPromM2 = topPromM2.reset_index()
```

```python
sns.set_theme()
plt.figure(figsize = (12,6))

sns.barplot(data = topPromM2, y = 'mean', x='Barrio')

sigma = topPromM2['std']

x = topPromM2['Barrio']
y = topPromM2['mean']

plt.errorbar(x = x, y = y, yerr=(sigma, sigma), fmt='none', color='k', capsize=5)

plt.title('Costo por m2 de inmuebles por barrio')
plt.xlabel('Barrio')
plt.ylabel('Promedio')

plt.xticks(rotation = 80)
plt.savefig('/content/drive/MyDrive/Inmersion-de-Datos-Alura/costoPorM2', format = 'png')
plt.show()
```

Primero de creo la columna con el valor por m2

Para después agrupar y usar describe para obtener el promedio y la desviación estándar

Luego se generaría el grafico. Se uso barplot y también se agregaron barras de error usando la desviación estándar.

## Clase 03 - Profundizando el análisis exploratorio

```python
inmuebles.groupby('Barrio').mean()
```

Para profundizar más en los datos se han agrupado todos los barrios de la ciudad para reflejar el promedio de todas las columnas numéricas de todos los barrios.

```python
datosBarrio = inmuebles.groupby('Barrio').sum()
datosBarrio
```

Ahora tomamos todos los datos por barrio y sumamos cada uno de sus atributos y con base a estos datos se obtendrá a continuación el promedio.



```python
datosBarrio['Valor_m2_Barrio'] = datosBarrio['Precio_Millon']/datosBarrio['Area']
datosBarrio
```

Aquí obtenemos el valor por metro cuadrado con respecto al promedio del propio barrio que es el resultado de la suma de todos los precios de un barrio y dividirlo entre los metros cuadrados de ese barrio.

```python
dict(datosBarrio['Valor_m2_Barrio'])
```

Procedente de esto se genera un diccionario donde se almacenan los nuevos valores.

```python
m2Barrio = dict(datosBarrio['Valor_m2_Barrio'])
```

Estos datos se almacenan en una nueva variable.

```python
inmuebles['Valor_m2_Barrio'] = inmuebles['Barrio']
inmuebles['Valor_m2_Barrio'] = inmuebles['Valor_m2_Barrio'].map(m2Barrio)
inmuebles.head(5)
```

Los datos de este diccionario son una llave que son asignados a una nueva columna en el dataset teniendo una nueva referencia la cual es el valor por metro cuadrado de cada inmueble.

```python
inmuebles['Barrio'].value_counts()[:10].index
```

Esta instrucción nos muestra una serie donde podemos observar a los 10 barrios con más inmuebles en venta.

```python
topBarrios = inmuebles['Barrio'].value_counts()[:10].index
```

Con base en estos datos se crea una variable que contiene los índices.

```python
datosBarrio.reset_index(inplace=True)
datosBarrio
```

Con esta instrucción se resetea barrios y se convierte en una columna.

```python
datosBarrio.query('Barrio in @topBarrios')
```

Esta es una consulta que devuelve los datos de los 10 barrios con más inmuebles en venta.

```python
plt.figure(figsize=(13,8))
ax = sns.barplot(x='Barrio', y='Valor_m2_Barrio', data = datosBarrio.query('Barrio in @topBarrios'))
ax.tick_params(axis='x', rotation=45)
```

Generamos un gráfico con los datos de la consulta.

```python
plt.figure(figsize=(13,8))
ax = sns.boxplot(x='Barrio', y='Valor_m2_Barrio', data = datosBarrio.query('Barrio in @topBarrios'))
ax.tick_params(axis='x', rotation=45)
plt.show()
```

Aquí generamos un gráfico conocido como boxplot, pero con los parámetros del area.

```python
plt.figure(figsize=(10,8))
ax = sns.boxplot(x='Barrio', y='Precio_Millon', data = inmuebles.query('Barrio in @topBarrios'))
ax.tick_params(axis='x', rotation=45)
plt.show()
```

Este tipo de gráfico se observan valores mínimos y máximos lo que indica que porcentaje de los valores por metro cuadrado de los precios por millón de cada barrio.

```python
plt.figure(figsize=(13,7))
ax = sns.boxplot(x='Barrio', y='Precio_Millon', data = inmuebles.query('Barrio in @topBarrios & Precio_Millon < 1500'))
ax.tick_params(axis='x', rotation=45)
plt.show()
```

Al agregar la condición que 'Precio_Millon' sea menor que 1500 hace que se eliminen los valores inexistentes de nuestro gráfico y poderlo apreciar mejor.

```python
plt.figure(figsize=(11,7))
ax = sns.boxplot(x='Barrio', y='Area', data = inmuebles.query('Barrio in @topBarrios & Area < 500'))
ax.tick_params(axis='x', rotation=45)
plt.show()
```

Aplicamos el mismo gráfico, pero ahora con el 'Area'

Los gráficos de cajas permiten comparar dentro de una misma variable diferentes conjuntos de datos para obtener una conclusión más eficiente sobre la informacion que se está manejando.

```python
X = datosML[['COD_UPZ_GRUPO','Habitaciones','Banos','CONJUNTO_CERRADO','SALARIO_ANUAL_MI','TIENE_ESCRITURA']]

y = datosML["Precio_Millon"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 99)
modelo_1 = LinearRegression()
modelo_1.fit(X_train, y_train)
y_predictTest = modelo_1.predict(X_test)
y_predict_train = modelo_1.predict(X_train)
mae_test = mean_absolute_error(y_test, y_predictTest)
r2_test = r2_score(y_test, y_predictTest)
mae_train = mean_absolute_error(y_train, y_predict_train)
r2_train = r2_score(y_train, y_predict_train)
print(mae_test,r2_test)
print(mae_train,r2_train)
```

Vamos a iterar nuevamente, pero en vez de seleccionar un parámetro vamos a agregar 2 más para imprimir los datos de prueba aplicándolos a los datos de entrenamiento.

En los resultados podemos observar una mejoría en el desempeño de los modelos.



```python
X = datosML[['COD_UPZ_GRUPO','Habitaciones','Banos','CONJUNTO_CERRADO','SALARIO_ANUAL_MI','TIENE_ESCRITURA']]

y = datosML["Precio_Millon"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 99)
modelo_1 = LinearRegression()
modelo_1.fit(X_train, y_train)
y_predictTest = modelo_1.predict(X_test)
y_predict_train = modelo_1.predict(X_train)
mae_test = mean_absolute_error(y_test, y_predictTest)
r2_test = r2_score(y_test, y_predictTest)
mae_train = mean_absolute_error(y_train, y_predict_train)
r2_train = r2_score(y_train, y_predict_train)
print(mae_test,r2_test)
print(mae_train,r2_train)
```
